```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
# Load & Merge
df = pd.read_csv('/content/ml_dataset/train.csv')
meal = pd.read_csv('/content/ml_dataset/meal_info.csv')
center = pd.read_csv('/content/ml_dataset/fulfilment_center_info.csv')

df = df.merge(meal, on='meal_id')
df = df.merge(center, on='center_id')
```

```python
# Preprocessing
df['week'] = pd.to_datetime(df['week'])
df['year'] = df['week'].dt.year
df['month'] = df['week'].dt.month

df = df.drop(['week'], axis=1)

df = pd.get_dummies(df)
```

```python
# Inspect df columns to identify the correct time-related column
print(df.columns)
```

```python
# Split
X = df.drop('num_orders', axis=1)
y = df['num_orders']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)
```

```python
# Scaling (important for SVR)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model
model = SVR(kernel='rbf')
model.fit(X_train, y_train)
```
