```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Load dataset
df = pd.read_csv('/content/ml_dataset/train.csv')
meal = pd.read_csv('/content/ml_dataset/meal_info.csv')
center = pd.read_csv('/content/ml_dataset/fulfilment_center_info.csv')

# Merge datasets
df = df.merge(meal, on='meal_id')
df = df.merge(center, on='center_id')

df.head()
```

```python
# Preprocessing

df['year'] = df['week'].dt.year
df['month'] = df['week'].dt.month

df = df.drop(['week'], axis=1)

df = pd.get_dummies(df)
```

```python
# Split
X = df.drop('num_orders', axis=1)
y = df['num_orders']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)
```

```python
# Model
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
```

```python
# Predict
pred = model.predict(X_test)
```

```python
# Evaluate
print("Random Forest MAE:", mean_absolute_error(y_test, pred))
```
